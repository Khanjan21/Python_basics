```python
# 2) Describe the role of predefined keywords in Python and provide examples of how they are used in a
#    program

'''

1) if, else , elif
Used for conditional statements.
'''
#Example:-

x = 10
if x > 5:
    print("x is greater than 5")
elif x == 5:
    print("x is equal to 5")
else:
    print("x is less than 5")
    
'''
2) for,while,break,continue
Used for loops and controlling loop execution.
'''

# for loop
for i in range(5):
    if i == 3:
        break   
    print(i)

# while loop
count = 0
while count < 5:
    count += 1
    if count == 3:
        continue  
    print(count)

'''
3)True, False, None
Used to represent boolean values and the absence of a value.
'''

is_active = True
is_admin = False
result = None

if is_active:
    print("The user is active")
```

```python
# 3)Compare and contrast mutable and immutable objects in Python with examples

'''
1)Mutable Objects

Definition: Mutable objects can be changed after they are created. 
You can modify their content without changing their identity.

Examples:

1)Lists
2)Dictionaries
3)Sets

Characteristics:

Operations that modify the object do not create a new object.
They have methods that change the content in place.
'''
#1)List

my_list = [1, 2, 3]
print(my_list)  

my_list.append(4)
print(my_list) 

#Dictionary:
my_dict = {'a': 1, 'b': 2}
print(my_dict)  

my_dict['c'] = 3
print(my_dict)      



'''
Immutable Objects
Definition: Immutable objects cannot be changed after they are created. 
Any operation that modifies an immutable object will create a new object. 

Common Examples:
Integers
Floats
Strings
Tuples

Characteristics:
Operations that seem to modify the object actually create a new object.
They do not have methods that change the content in place (since itâ€™s not possible).
'''
#Examples:

#String:
my_string = "hello"
print(my_string)   

'''
my_string.append("world")
ERROR
'''
my_string = my_string + " world"
print(my_string)

#Tuple:
my_tuple = (1, 2, 3)
print(my_tuple)

'''
my_tuple.append(4)
ERROR
'''
my_tuple = my_tuple + (4,)
print(my_tuple)      # Output: (1, 2, 3, 4)
```

```python
#5)Explain the concept of type casting in Python with examples

'''
1. Implicit Type Casting
Python automatically converts one data type to another when it is required. 
This is called implicit type casting.
'''

a = 5        
b = 2.0       
c = a + b     
print(c)      
print(type(c))

'''
2) Explicit Type Casting
Explicit type casting is done manually by using Python's built-in functions.
'''

#Examples:

#Converting String to Integer:

s = "123"
num = int(s)
print(num)         
print(type(num)) 

#Converting Integer to String:
num = 123
s = str(num)
print(s)        
print(type(s))  

#Converting String to Float:
s = "123.45"
num = float(s)
print(num)         
print(type(num))   
```

```python
#6) How do conditional statements work in Python? Illustrate with examples
'''
1) if Statement
The if statement evaluates a condition and executes the indented block of code
that follows it if the condition is True.
'''
'''
Syntax:
if condition:
    # block of code
'''
#Example:
age = 18
if age >= 18:
    print("You are eligible to vote.")
 
'''
elif Statement
The elif (short for "else if") statement allows you to check multiple conditions sequentially. If the preceding if condition is False, the elif condition is evaluated. If the elif condition is True, its block of code is executed.

Syntax:

 
if condition1:
    # block of code
elif condition2:
    # block of code
 '''
 
score = 85
if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
else:
    print("Grade: D")

    '''
else Statement
The else statement follows an if or elif block and executes if none of the preceding conditions are True.

Syntax:

 
if condition1:
    # block of code
elif condition2:
    # block of code
else:
    # block of code
'''
#Example:
temperature = 30
if temperature > 30:
    print("It's a hot day.")
elif temperature < 10:
    print("It's a cold day.")
else:
    print("The weather is moderate.")

'''
Nested Conditional Statements
You can nest conditional statements within each other to handle more 
complex conditions.
'''
 
num = 10
if num >= 0:
    if num == 0:
        print("Zero")
    else:
        print("Positive number")
else:
    print("Negative number")
```

```python
# 7) Describe the different types of loops in Python and their use cases with examples.
'''
1. for Loop
The for loop is used to iterate over a sequence
(such as a list, tuple, dictionary, set, or string) or 
other iterable objects. It executes a block of code for each 
item in the sequence.

Syntax:
for variable in sequence:
    # block of code
    '''
#Examples:

#Iterating Over a List:
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
    
#Using range():
#The range() function generates a sequence of numbers,
#which is commonly used with for loops.
for i in range(5):
    print(i)
    
#Iterating Over a Dictionary:
student_grades = {"Khanjan": 90, "Ajay": 85, "Bijay": 92}
for student, grade in student_grades.items():
    print(f"{student}: {grade}")
 

'''
2. while Loop
The while loop is used to execute a block of code as long as a given condition is True. It is typically used when the number of iterations is not known beforehand.

Syntax:

 
while condition:
    # block of code
    '''
#Examples:
count = 0
while count < 5:
    print(count)
    count += 1
    
#Using while Loop for User Input:
user_input = ""
while user_input.lower() != "quit":
    user_input = input("Enter something (type 'quit' to stop): ")
    print(f"You entered: {user_input}")
    
#Infinite Loop (with a break condition):
while True:
    name = input("Enter your name (type 'exit' to stop): ")
    if name.lower() == "exit":
        break
    print(f"Hello, {name}!")
'''  
3. Control Statements in Loops
Examples:
'''

#Using break:
for num in range(10):
    if num == 5:
        break
    print(num)
    
#Using continue:

 
for num in range(10):
    if num % 2 == 0:
        continue
    print(num)
```
